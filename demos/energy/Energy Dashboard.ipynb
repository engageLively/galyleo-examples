```python
def break_out_national(df):
    national = df.loc[df['State'] == 'United States']
    all_states = df.loc[(df['State'] != 'United States')]
    only_states = all_states.loc[all_states['State'] != 'District of Columbia']
                        
    return (only_states, national )
(state_data, national_data) = break_out_national(df)
(state_production, national_production) = break_out_national(production)
(state_consumption, national_consumption) = break_out_national(production)
```

```python
from galyleo.galyleo_table import GalyleoTable
from galyleo.galyleo_jupyterlab_client import GalyleoClient
```

```python
def send_dataframe_to_dashboard(dataframe, table_name, client):
    table = GalyleoTable(table_name)
    table.load_from_dataframe(dataframe)
    client.send_data_to_dashboard(table)
```

```python
client = GalyleoClient()
```

```python
send_dataframe_to_dashboard(state_data, 'State Data', client)
send_dataframe_to_dashboard(national_data, 'National Data', client)
send_dataframe_to_dashboard(state_production, 'State Production', client)
send_dataframe_to_dashboard(national_production, 'National Production', client)
send_dataframe_to_dashboard(state_production, 'State Consumption', client)
send_dataframe_to_dashboard(national_production, 'National Consumption', client)
```
